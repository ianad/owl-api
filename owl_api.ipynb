```python
df_hero_baseline = pd.DataFrame(columns=['hero_name','damage','eliminations','healing', 'deaths'])

for hero_name,map_game_type in itertools.product(player_hero_thresholds['hero_name'].unique(),
                                       player_hero_thresholds['map_game_type'].unique()):
  
  uid = player_hero_thresholds.loc[(player_hero_thresholds['hero_name']==hero_name)&
                           (player_hero_thresholds['map_game_type']==map_game_type), 
                           'esports_player_id'].unique()
  
  _df_hero_baseline = (df_player_map_data
                   .loc[(df_player_map_data['esports_player_id']
                         .isin(uid))&(df_player_map_data['hero_name']==hero_name), 
                        ['damage','eliminations','healing', 'deaths']]
                   .mean()
                   .to_frame().T.reset_index()
                   .rename(columns={'index':'dist'}))
  
  _df_hero_baseline['dist'] = 'mean'
  
  for q in [0.01,0.1,0.25,0.5,0.75,0.9,0.99]:
    _df = pd.DataFrame()
    _df = (_df
           .append(df_player_map_data
                   .loc[(df_player_map_data['esports_player_id']
                         .isin(uid))&(df_player_map_data['hero_name']==hero_name), 
                        ['damage','eliminations','healing', 'deaths']]
                   .quantile(q)
                   .to_frame().T.reset_index()
                   .rename(columns={'index':'dist'}), 
                   ignore_index=True))
    _df_hero_baseline = _df_hero_baseline.append(_df,ignore_index=True,sort=True) 
    _df_hero_baseline['hero_name'] = hero_name
    _df_hero_baseline['map_game_type'] = map_game_type
  df_hero_baseline = df_hero_baseline.append(_df_hero_baseline,ignore_index=True,sort=True)
df_hero_baseline = df_hero_baseline.merge(df_hero_roles)  
```

```python
_matches = pd.Series(w_die['esports_match_id'].unique())
df_pred = pd.DataFrame()
kf = KFold(10,shuffle=True)
for train_idx,test_idx in kf.split(_matches):
  
  test_games = _matches[test_idx]
  test_data = w_die.loc[w_die['esports_match_id'].isin(test_games)].dropna()
  train_data = w_die.loc[~w_die['esports_match_id'].isin(test_games)].dropna()
  
  logr = LogisticRegression(solver='lbfgs')
  logr.fit(train_data.loc[:,['deaths']],train_data['w'])
  test_data['pred'] = pd.DataFrame(logr.predict_proba(test_data.loc[:,['deaths']]),index=test_data.index)[1]
  df_pred = df_pred.append(test_data)
```
